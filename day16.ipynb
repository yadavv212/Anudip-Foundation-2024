```python
#Example:Visualize the number of books sold in a bookstore by genre over a year

import matplotlib.pyplot as plt
# Sample data: Number of books sold by genre over a year
genres = ["Mystery", "Romance", "Science Fiction", "Fantasy", "Thriller"]
books_sold = [120, 90, 80, 110, 70]
# Create the bar plot
plt.bar(genres, books_sold)
# Add labels and a title
plt.xlabel("Genre")
plt.ylabel("Number of Books Sold")
plt.title("Books Sold by Genre Over a Year")
# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Sample data: Population distribution by age group in a city
age_groups = ["0-10", "11-20", "21-30", "31-40", "41-50", "51-60", "61-70", "71+"]
population = [15000, 22000, 30000, 28000, 25000, 18000, 12000, 8000]
# Create the bar plot
plt.barh(age_groups, population, color='lightblue')
# Add labels and a title
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.title("Population Distribution by Age Group")
# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Sample data: Population distribution by age group in a city
age_groups = ["0-10", "11-20", "21-30", "31-40", "41-50", "51-60", "61-70", "71+"]
population = [15000, 22000, 30000, 28000, 25000, 18000, 12000, 8000]
# Create the bar plot
plt.bar(age_groups, population, color='green',width=0.5,alpha=0.7,edgecolor="m",linewidth=5,linestyle=":",label="Population")
# Add labels and a title
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.title("Population Distribution by Age Group")
plt.legend()
# Display the plot
plt.show()
```

```python
##practice question
import matplotlib.pyplot as plt

# Example list of exam scores
exam_scores = [22, 45, 60, 75, 80, 50, 55, 70, 85, 90, 40, 35, 65, 77, 82, 45, 50, 68, 73, 78]

# Create the histogram
plt.hist(exam_scores, bins=10, edgecolor='black')

# Add titles and labels
plt.title('Distribution of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Number of Students')

# Show the plot
plt.show()
```

```python
#piechart
import matplotlib.pyplot as plt
# Market share data
manufacturers = ['Samsung', 'Apple', 'Huawei', 'Xiaomi', 'Others']
market_share = [30, 25, 18, 12, 15]
# Colors for each slice
colors = ['lightblue', 'lightcoral', 'lightgreen', 'lightpink', 'lightgrey']
# Create a pie chart
plt.pie(market_share, labels=manufacturers, colors=colors, autopct='%1.f%%')
# Title
plt.title('Smartphone Market Share')
# Show the pie chart
plt.show()
```

```python
#scatter plot
import matplotlib.pyplot as plt
# Student data (study hours and exam scores)
study_hours = [2, 3, 1, 4, 3, 5, 2, 6, 5, 7]
exam_scores = [65, 75, 60, 80, 70, 85, 70, 90, 88, 92]
# Create a scatter plot
plt.scatter(study_hours, exam_scores, c='green', marker='o', label='Student Data')
# Labeling and Title
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Study Hours vs. Exam Scores')
# Show the scatter plot
plt.legend()
plt.grid(True)
plt.show()
```
